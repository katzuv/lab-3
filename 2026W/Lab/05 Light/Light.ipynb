### Diffraction from a single slit
Assemble the experimental setup as shown in Figure 13. Ensure that the beam remains collimated after passing through the two lenses. Collimation means that the size of the beam should not change significantly for different distances.If the beam is not coolimated, adjust the distance between the lenses. Position the single slit in the path of the beam and observe the interference pattern. Adjust the width of the slit to observe a distinct diffraction pattern.

The theoretical expression for the slit width is

$$
a\sin \theta_n = \pm n\lambda,n=1,2,3\dots
$$

Where $$\theta_n = \arctan \left(\frac{X_{\text{0 order-minima number n}}}{L}\right)$$ with L being the distance between the slit and the screen.

```python
## Does the result match the width you selected for the slit? If not, please explain why here.
measured_a = .75 / 1000
L = 1.3
lam = 532e-9
x0 = 1.25e-3
theta = np.arctan(x0 / L)
a = lam / np.sin(theta)
print(a * 1000)
(measured_a - a) / a
```
